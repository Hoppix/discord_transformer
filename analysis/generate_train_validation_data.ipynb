```python
import pandas as pd
import random
```

## Load all the data frames and merge them into one

```python
chat_df = pd.DataFrame()
with open("../data/names.txt", "r+") as file:
    for name in file:
        name = name.strip()
        df = pd.read_csv(f"../data/{name}_messages.csv")
        chat_df = pd.concat([chat_df, df])
chat_df.to_csv("../data/merged_messages.csv")
chat_df
```

## Parse all the sentences by a single user

```python
USER = "Hoppix#6723"
chat_df = df.loc[df["Author"] == "Hoppix#6723"]
sentences = chat_df["Content"].tolist()
chat_df = chat_df.loc[chat_df["Author"] == USER]
sentences = chat_df["Content"].tolist()
sentences = [ str(sentence) for sentence in sentences]
len(sentences)
```

## Shuffle the entries

```python
random.shuffle(sentences)
sentences
```

## Save training data

```python
with open("../data/training_data.csv", "w+", encoding="utf-8") as file:
    for sentence in sentences:
        file.write(str(sentence) + "\n")
```
